```python
import torch

model.model.load_state_dict(torch.load("./yolov8n.pt")["model"].state_dict())
```

```python
result = model.predict("./test.jpg", classes=[0,])
```

```python
result[0].save("./result.jpg")
```

```python
import cv2

vid = cv2.VideoCapture("test_video.mp4")

key = 0
while key != 27:
    ret, frame = vid.read()
    if ret:
        cv2.imshow("Frame", frame)
        preds = model.track(frame, classes=[0,], verbose=False, conf=0.5)
        for r in preds:
            im_array = r.plot()  # plot a BGR numpy array of predictions
            cv2.imshow("preds", im_array)
    key = cv2.waitKey(100)
vid.release()
cv2.destroyAllWindows()
```
